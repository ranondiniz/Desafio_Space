# Carregamento e Tratamentos:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# space_DF = Space_Corrected.csv('Space_Corrected.csv', header=0)
```

```python
# visualização da amostra de dados

space_DF = pd.read_csv('Space_Corrected.csv', header=0)
```

```python
# calculando as estatísticas descritivas para o DF
stats = space_DF.describe().T
print(stats)
```

```python
# exclusão de colunas desnecessárias

space_DF = space_DF.drop('Unnamed: 0', axis = 1)
space_DF = space_DF. drop('Unnamed: 0.1', axis = 1)
```

```python
space_DF.head(30)
```

```python
# verificando informações gerais das estatísticas após exclusão de colunas

space_DF.describe()
```

```python
# verificando a existência e quantidade de dados nulos na base de dados

space_DF.isnull().sum()
```

```python
# removendo registros com valores nulos da coluna

space_DF = space_DF.dropna(subset=[' Rocket'])
print(space_DF)
```

```python
# verificando se a coluna ainda possui dados nulos

print(space_DF[' Rocket'].isna().sum())
```

```python
# criando a coluna País a partir de Location

coun_func = lambda x:x.split(',')[-1]
space_DF['Country'] = space_DF['Location'].apply(coun_func)
```

```python
space_DF.head(7)
```

```python
# criando as colunas Mês e Ano a partir da coluna Datum

month_function = lambda x:x.split( )[1]
space_DF['Month'] = space_DF['Datum'].apply(month_function)

year_function = lambda x:x.split( )[3]
space_DF['Year'] = space_DF['Datum'].apply(year_function)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# qual é o total de missões registradas?
contagem_missoes = space_DF['Status Mission'].value_counts()

# soma dos resultados
total_missoes = contagem_missoes.sum()

contagem_missoes['Total'] = total_missoes

# Imprimir a contagem geral das categorias
#print("Contagem Geral das Missões:")
#print(contagem_missoes)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=contagem_missoes.index, y=contagem_missoes.values, palette=['#66b3ff'] * len(contagem_missoes))

# visualizar o número nas barra
for i, valor in enumerate(contagem_missoes.values):
    ax.text(i, valor + 0.5, f'{valor}', ha='center', color='black')

plt.title('Contagem Total de Missões Registradas')
plt.xlabel('Status da Missão')
plt.ylabel('Número de Missões')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# criando o countplot da coluna
ax = sns.countplot(data=space_DF, x='Status Mission')

plt.title('Contagem de Missões por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Total de Missões')

plt.xticks(rotation=45)

# visualizar o número nas barra
total_missions = len(space_DF)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height, '{:.1f}%'.format((height / total_missions) * 100),
            ha="center", va="bottom", fontsize=10, color="black")
    p.set_color('#66b3ff')

plt.show()
```

# Dados Interessantes:

```python
# qual o total de missões agrupadas por companhia e país?

# agrupamento de 'Company Name' e 'Country' para contar as missões
missoes_por_empresa_pais = space_DF.groupby(['Company Name', 'Country']).size().reset_index(name='Total Missões')

# print(missoes_por_empresa_pais)

# gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x='Total Missões', y='Company Name', hue='Country', data=missoes_por_empresa_pais, dodge=False)

plt.xlabel('Total de Missões')
plt.ylabel('Empresa')
plt.title('Total de Missões por Empresa e País')
plt.legend(title='País')
plt.tight_layout()

plt.show()
```

```python
# qual o total de missões bem-sucedidas agrupadas por companhia e país?

# agrupamento de 'Company Name', 'Country', e 'Status Mission' para contar as missões
missoes_status = space_DF.groupby(['Company Name', 'Country', 'Status Mission']).size().reset_index(name='Total Missões')

# print(missoes_status)

# gráfico de barras
plt.figure(figsize=(14, 10))
sns.barplot(x='Total Missões', y='Company Name', hue='Status Mission', data=missoes_status, dodge=True)

plt.xlabel('Total de Missões')
plt.ylabel('Empresa')
plt.title('Total de Missões por Empresa e Status')
plt.legend(title='Status da Missão')
plt.tight_layout()

plt.show()
```

```python
# quais as 3 empresas e 3 países que mais tiveram sucesso nas suas missões?

# filtro para pegar as missões bem-sucedidas
missoes_sucesso = space_DF[space_DF['Status Mission'] == 'Success']

# agrupamento por 'Company Name'
empresa_sucesso = missoes_sucesso.groupby('Company Name').size().reset_index(name='Total Missões')
top_empresas = empresa_sucesso.nlargest(5, 'Total Missões')

# agrupamento por 'Country'
pais_sucesso = missoes_sucesso.groupby('Country').size().reset_index(name='Total Missões')
top_paises = pais_sucesso.nlargest(5, 'Total Missões')

# filtro para pegar apenas as top empresas e países
top_empresas_lista = top_empresas['Company Name'].tolist()
top_paises_lista = top_paises['Country'].tolist()

missoes_filtradas = space_DF[(space_DF['Company Name'].isin(top_empresas_lista)) & (space_DF['Country'].isin(top_paises_lista))]

# agrupamento por empresa, país, e status da missão para contagem
missoes_agrupadas = missoes_filtradas.groupby(['Company Name', 'Country', 'Status Mission']).size().reset_index(name='Total Missões')

#print(missoes_agrupadas)

plt.figure(figsize=(14, 10))

g = sns.catplot(
    x='Company Name',
    y='Total Missões',
    hue='Status Mission',
    col='Country',
    data=missoes_agrupadas,
    kind='bar',
    height=6,
    aspect=1,
    col_wrap=3,
    dodge=True
)

g.set_axis_labels("Empresa", "Total de Missões")
g.set_titles("País: {col_name}")
g.fig.suptitle('Total de Missões por Empresa, País e Status (Top 3)', y=1.03)

# separação entre os gráficos
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.show()
```

```python
# qual a porcentagem de lançamentos para cada mês?

# verificando se os dados estão carregados corretamente
print(space_DF['Month'].unique())

# ordenação dos meses
meses_ordenados = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

lançamentos_por_mes = space_DF['Month'].value_counts().reindex(meses_ordenados)

# calculando a porcentagem de lançamentos por mês
total_lancamentos = lançamentos_por_mes.sum()
porcentagem_por_mes = (lançamentos_por_mes / total_lancamentos) * 100

plt.figure(figsize=(12, 8))
ax = sns.barplot(x=porcentagem_por_mes.index, y=lançamentos_por_mes.values, palette='viridis')

# visualizar a porcentagem nas barras
for i, porcentagem in enumerate(porcentagem_por_mes):
    ax.text(i, lançamentos_por_mes[i] + 0.5, f'{porcentagem:.1f}%', ha='center', color='black')

plt.title('Sazonalidade dos Lançamentos Espaciais por Mês')
plt.xlabel('Mês')
plt.ylabel('Número de Lançamentos')
plt.xticks(rotation=45)
plt.show()
```

```python
# qual a porcentagem de lançamentos para cada ano?

lançamentos_por_ano = space_DF['Year'].value_counts().sort_index()

# calculando a porcentagem de lançamentos por ano
total_lançamentos = lançamentos_por_ano.sum()
porcentagem_por_ano = (lançamentos_por_ano / total_lançamentos) * 100

plt.figure(figsize=(12, 8))
bars = plt.bar(porcentagem_por_ano.index, porcentagem_por_ano.values, color='skyblue')

# visualizar a porcentagem nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom')

plt.title('Porcentagem de Lançamentos por Ano')
plt.xlabel('Ano')
plt.ylabel('Porcentagem de Lançamentos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
